````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

prev_renda = pd.read_csv('previsao_de_renda.csv')
prev_renda.head()
````

````python
# Separando as variaveis que serão utilizadas
prev_renda_quant = prev_renda[['renda','posse_de_veiculo','posse_de_imovel']]
````

````python
# Avaliando com um grafico pointplot para posse de veiculo
sns.pointplot(x = 'posse_de_veiculo', y ='renda',
             data = prev_renda_quant, dodge = True, errorbar=('ci', 90))
````

````python
# Avaliando com um grafico pointplot para posse de imovel
sns.pointplot(x = 'posse_de_imovel', y ='renda',
             data = prev_renda_quant, dodge = True, errorbar=('ci', 90))
````

````python
# Avaliando com um grafico pointplot para posse de veiculo
sns.pointplot(x = 'posse_de_veiculo', y ='renda', hue = 'posse_de_imovel',
             data = prev_renda_quant, dodge = True, errorbar=('ci', 95))
````

- Podemos ver que quem possuem renda maior que 4600 costuma possuir imovel e automovel, isso pode ser significativo para identificarmos uma possivel autorização de credito.
