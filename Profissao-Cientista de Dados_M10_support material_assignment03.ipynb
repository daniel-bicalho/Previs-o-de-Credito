1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Convertendo a coluna booleana para um tipo string para poder usar o metodo 'startswith'
prev_renda['posse_de_imovel'] = prev_renda['posse_de_imovel'].astype('str')

# Criação do count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=prev_renda, x='data_ref', hue='posse_de_imovel')

# Adicionando título e rótulos
plt.title('Posse de Imovel por Data')
plt.xlabel('Data de Referência')
plt.ylabel('Contagem')
plt.legend(title='Possui Imovel')
plt.xticks(rotation=90)
plt.tight_layout()
````

2) Para cada variável quantitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x="data_ref", y="qtd_filhos",
                   data=prev_renda, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = prev_renda['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(prev_renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````
