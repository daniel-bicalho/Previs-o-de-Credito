```python
# import math

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

# from sklearn.model_selection import train_test_split
# from sklearn import metrics
# from sklearn.ensemble import RandomForestClassifier

# from scipy.stats import ks_2samp
import statsmodels.formula.api as smf
import statsmodels.api as sm
import patsy

%matplotlib inline
```

```python
df = pd.read_csv('previsao_de_renda.csv')
df.info()
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.  


2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.  


3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python
# Ajustando um modelo de regressão linear múltipla no log de renda
reg = smf.ols('np.log(renda) ~ tempo_emprego + qt_pessoas_residencia + idade + tipo_residencia + estado_civil + educacao  + tipo_renda + qtd_filhos + sexo + posse_de_veiculo + posse_de_imovel',
              data = df).fit()
reg.summary()
```

```python
# tratando as variaveis qualitativas como dummies
reg = smf.ols('np.log(renda) ~ tempo_emprego + qt_pessoas_residencia + idade + C(tipo_residencia) + C(estado_civil) + C(educacao)  + C(tipo_renda) + qtd_filhos + C(sexo) + C(posse_de_veiculo) + C(posse_de_imovel)',
              data = df).fit()
reg.summary()
```

```python
# Buscando a casela de referencia - teste sexo M
y, x = patsy.dmatrices('np.log(renda) ~ tempo_emprego + qt_pessoas_residencia + idade + C(tipo_residencia) + C(estado_civil) + C(educacao) + C(tipo_renda) + qtd_filhos + C(sexo, Treatment("M")) + C(posse_de_veiculo) + C(posse_de_imovel)',
              data = df)
sm.OLS(y, x).fit().summary()
```

- Não tivemos muita diferença vamos tentar retirar as variaveis sem muita coorrelação

```python
# Remova as variáveis não significantes e ajustando novamente o modelo
reg = smf.ols('np.log(renda) ~ tempo_emprego + idade + tipo_renda  + sexo + educacao + posse_de_imovel',
              data = df).fit()
reg.summary()
```

```python
# Buscando a casela de referencia para melhorar o p - value
reg = smf.ols('np.log(renda) ~ tempo_emprego + idade + C(tipo_renda, Treatment("Empresário"))  + C(sexo, Treatment("M")) + C(educacao, Treatment("Pós graduação")) + C(posse_de_imovel)',
              data = df).fit()
reg.summary()
```

- Conseguimos melhorar o p-value do nosso modelo 
- mas o valor de R2 segue o mesmo vamos retirar todas as variaveis com o valor de R2 acima de 5% e comparar os modelos.

```python
reg = smf.ols('np.log(renda) ~ tempo_emprego + idade  + C(sexo, Treatment("M")) + C(educacao, Treatment("Pós graduação")) + C(posse_de_imovel)',
              data = df).fit()
reg.summary()
```

- temos um modelo com o melhor p-value oque significa variaveis explicativas melhores e mesmo o R2 sendo menor temos o F-statistic significativamente maior, assim o modelo ficou com um desempenho melhor em termos de força explicativa.
